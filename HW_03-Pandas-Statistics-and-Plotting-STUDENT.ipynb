## Part 2: Datasaurus, or the perils of summary statistics (34 total points)

In the 1970s, the statistician F. J. Anscombe published the article ["Graphs in Statistical Analysis"](https://www.jstor.org/stable/2682899), which demonstrated the importance of actually graphing your data, instead of just looking at the summary statistics: mean, median, inter-quartile ranges, etc.  

To hammer this point home, he created "Anscombe's quartet", a group of four graphs that are completely different, yet have the same means, ranges, and lines of linear regression.

If the prior steps were done correctly, the following code will take the information in the dataframe and convert it to another format for easier access:

````python
data_dict = {}
for dataset_name in datasets:
    data_dict[dataset_name] = {}
    data_dict[dataset_name]['x'] = data['x'][data['dataset'] == dataset_name].values  
    data_dict[dataset_name]['y'] = data['y'][data['dataset'] == dataset_name].values  
````

**Structure:** Dictionary > *[Key]* Dataset Name > *[Value]* Nested Dictionary > *[Nested Key]* x or y > *[Nested Value]* Array with x or y values for the given dataset

Dictionary slicing using keys can be done to access data. Two slicing operations need to be done in order to access x and y values, since there are nested dictionaries involved. Here is a MWE to access x and y values of the data.

### 2.2: Getting summary statistics (11 points)

Now let's compute the following statistics for each dataset:

- mean of x coordinates
- mean of y coordinates
- standard deviation of x coordinates
- standard deviation of y coordinates
- correlation between x and y  (**Hint: check out `np.corrcoef`**)

&#9989; **Task 2.2.1: (9 points)** Make a loop that prints this information for all datasets in the following format:

```
Dataset: dino       x_mean = 54.26 | x_std = 16.71 | y_mean = 47.83 | y_std = 26.84 | corr = -0.06
Dataset: away       x_mean = 54.27 | x_std = 16.71 | y_mean = 47.83 | y_std = 26.84 | corr = -0.06
...
```

````python
from numpy import * # another way I learned to import modules - this way, I don't need to invoke the module name repeatedly, especially in longer lines of code

for dataset in data_dict: # accessing datasets in the dictionary via a loop by value
    # the if-else statement is only for the aesthetics and formatting. the code under both selection statements are logically the same.
    if len(dataset) == 4: 
        print("Dataset:", dataset, "\t\tx_mean =", round(mean(data_dict[dataset]["x"]), 2), # uses the round function to round to two decimal places
              "| x_std =", format(round(std(data_dict[dataset]["x"]), 2), '.2f'), # uses string formatting to ensure there are two decimal places when the second decimal place is 0.
              # Source: https://www.geeksforgeeks.org/python-string-format-method/; Accessed: 03/08/2024
              "| y_mean =", round(mean(data_dict[dataset]["y"]), 2), 
              "| y_std =", round(mean(data_dict[dataset]["y"]), 2), 
              "| corr =", round(corrcoef(data_dict[dataset]["x"], data_dict[dataset]["y"])[0][1], 2)) 
                # this needs additional slicing since the result is a nested list of two lists with correlation coefficient values
                # Source: https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html; Accessed: 03/08/2024
    else:
        print("Dataset:", dataset, "\tx_mean =", round(mean(data_dict[dataset]["x"]), 2), 
              "| x_std =", format(round(std(data_dict[dataset]["x"]), 2), '.2f'), 
              "| y_mean =", round(mean(data_dict[dataset]["y"]), 2), 
              "| y_std =", round(mean(data_dict[dataset]["y"]), 2), 
              "| corr =", round(corrcoef(data_dict[dataset]["x"], data_dict[dataset]["y"])[0][1], 2))
````
