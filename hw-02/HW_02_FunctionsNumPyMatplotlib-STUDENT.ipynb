```python
def evolved_sum_of_list(intlist, initial_value): # definition of function with two inputs, one for the list and the other for the initial value
    total = initial_value # declaring a variable to store the total, but having the inital value as the base value in the variable
    for value in intlist: # initializing a loop that loops by value in the list of integers
        total += value # using the increment operator to add each value with each iteration of the loop
    return total # returning the total
```

### &#9989;&nbsp; Question 1.4 (4 points)

Write a function to return the median of a list of integers.

Reminder: The median is computed by sorting the values from least to greatest. Then, if the number of values is odd, the median is the middle value. Otherwise, the median is the average of the middle two values.

Examples:
- The median of `[2, 1, 3]` is `2` since the sorted values are `[1, 2, 3]` and the middle value is `2`.
- The median of `[3, 4, 1, 2]` is `2.5` since the sorted values are `[1, 2, 3, 4]` and the average of the middle values is `(2 + 3) / 2 = 2.5`.

Note: Write the function "from scratch" - don't use any other functions to do this for you. You can, however, use the built-in [sorted](https://docs.python.org/3/howto/sorting.html) function to sort the values from least to greatest.

```python
def median(input_list): # defines function that takes a list as a parameter
    sorted_list = sorted(input_list) # sorts the list in ascending order
    if len(sorted_list) % 2 == 0: # checks if the length of the list is even by using the modulus operator
        median = average([sorted_list[int(len(sorted_list)/2) - 1], sorted_list[int(len(sorted_list)/2)]]) # uses the previously defined average function to calculate the mean value of the middle two numbers in the list, if the length is even
    else:
        median = sorted_list[int(len(sorted_list)/2)] # if the length is odd, slicing is used to extract the middle-most value
    return median # returns the median value

# https://docs.python.org/3/howto/sorting.html, Accessed: 02/08/2024
```

We will cover random numbers later in the semester, but for now all you need to know is that truly random numbers are almost impossible to create using computers. There are functions that create sequences of numbers that can mimic the properties of random numbers, but they are not truly random because the function will always output the same sequence everytime its called. 

If you want to know more about it now you can read [this website](https://numpy.org/doc/stable/reference/random/index.html) and/or ask your instructor.



```python
# Import the needed libraries
import numpy as np

# Create a pseudo-random number generator and call it rng. 
# The input 0 is the seed of the random number generator. This is needed because we want the result to be reproducible
rng = np.random.RandomState(0)
# Create an array of 10 random integers picked from the range (-5, 5). 
values = rng.randint(low=-5, high=5, size=10)

# Create an outlier, e.g. a number outside the original range.
outlier = 1000

# Your code here.

# calculating the average and median of the original list of values
print("Average of the original list:", average(values))
print("Median of the original list:", median(values))

# introducing the outlier to the list
values = np.append(values, outlier)

# calculating the new average and median
print("Average of the modified list:", average(values))
print("Median of the modified list:", median(values))
```

<font size=6 color="#009600">&#9998;</font>

*If one were to replace a single data point with an outlier in a sample, the mean of the sample would change drastically. This means that the sample mean is extremely sensitive to contamination, indicating that the sample mean is not a robust estimator. In statistics, the robustness of an estimator, like the mean, is measured using the finite-sample replacement breakdown point, the value of which for the mean is 0. On the other hand, one looks at the middle data point for the median, as opposed to the average value as in the case of the mean. The sample median is more robust to outliers compared to the sample mean. It is less affected by extreme values because it is based on the middle value of the sorted data, and not on the sum of all values. If less than half of the data points are contaminated, the sample median remains unaffected, or at least provides a reliable estimate of the central tendency of the data, which is a goal of the median, being a measure of central tendency. Therefore, the sample median has a higher RBP compared to the sample mean, a value of $\frac{n}{2}$, where $n$ is the sample size. This can be observed above, where the value of the mean changes by nearly a factor of 100, while in the case of the median, the change is very less, i.e., by 0.5 units.*

NumPy contains many mathematical functions that we use every day - for example `sin`, `cos`, `exp`, and many more. You can apply these functions to single numbers, but the real power of NumPy comes from the ability to apply these functions to many values at once.

In the cell below, you're provided with a list of floating point values. Your task is to:

1. Take the exponential of each value.
2. Compute the sum of all these values.
3. Divide the exponential of each value by the sum of all values, and store these results in a list.

Example:
- Given the list [$a$, $b$], the final list you end up with should be [$\frac{e^a}{e^a + e^b}$, $\frac{e^b}{e^a + e^b}$].

First do this using only the `math.exp` function from Python's math library (`import math`). This function only inputs a single number, not a list, and you'll have to use loops.

Then, do this using the `np.exp` function from NumPy which can input a list of numbers.

Print out your answers in both cases to show that they are the same.

*Hint*: The final lists in both cases should be all positive numbers that sum to `1.0`.

*Fun fact*: This sequence of operations is called a [softmax](https://en.wikipedia.org/wiki/Softmax_function) and it is often used in machine learning, among other applications.

```python
values = [-2.77174688,  0.65848257,  4.5938291, -1.04974511, -0.06412875, 2.09031183, -1.46996295,  1.32916018,  0.5168653, -0.2175936]

# Your code goes here.

# importing modules
import math
import numpy as np

# using the math module

values_new = [] # declaring a new list to store the modified values
for value in values: # initiating a loop to calculate the exponents and store them in the new list
    values_new.append(math.exp(value))
sum_values_new = 0
for value in values_new: # calculating the sum of the list
    sum_values_new += value

for i in range(len(values)): # dividing each value by the sum of the entire list using a loop by index in order to modify the list values instead of creating a new list. uses unary division assignment operator.
    values_new[i] /= sum_values_new

print(values_new)
print("Sum of all values:", sum(values_new))

# using numpy and arrays

values_array = np.array(values) # creates an array from the original list
values_array = np.exp(values_array)
sum_values_array = np.sum(values_array)
values_array = values_array/sum_values_array

print(values_array)
print("Sum of all values:", sum(values_array))
```

### &#9989;&nbsp; Question 2.3 (7 points)

There's a file on the course website called `random_lists.txt` which contains lists of random lists. The file has 20 rows and each row is a list of random numbers. Your task is to determine which lists are probability distributions, and convert the ones that are not probability distributions into probability distributions by using the softmax function from Question 2.1.

You should:
- Use NumPy to load in the `random_lists.txt` data.
- Define a function to return True if a list of values is a probability distribution, else return False. (A probability distribution is a sequence of non-negative numbers that sums to one.)
- Define a function to apply softmax (using NumPy).
- Loop through the `random_lists` and check if each list is a probability distribution. If it is, store the index. If it is not, apply a softmax to convert it to a probability distribution.
- Print out the indices of the lists that were probability distributions.

*Hint*: The list at index `1` in `random_lists` is a probability distribution. There may or may not be others.

```python
import numpy as np
data = np.loadtxt("random_lists.txt") # loading the data into a variable. loaded in the form of nested NumPy arrays.

def check_prob_dist(dataset): # defining a function to check if every distribution is a probability distribution. accepts a singular object as parameter, which is intended to be an array
    if np.isclose(np.sum(dataset), 1.0): # checks if the sum of the values in the array is close to 1.0
        return True
    else:
        return False

def softmax(dataset): # defines the softmax function, accepting a singular object as parameter, which is intended to be an array
    dataset = np.exp(dataset) # exponentiates all values in the dataset
    sum_data = np.sum(dataset) # calculates the sum of the exponentiated values
    dataset = dataset/sum_data # divides all the exponentiated values by the above sum to create a probability distribution
    return dataset

for i in range(len(data)): # loop by index
    if check_prob_dist(data[i]): # runs the check function on each individual distribution
        print("Data is a probability distribution:") # prints the distribution if true
        print(data[i], end = "\n\n")
    else:
        data[i] = softmax(data[i]) # converts to probability distribution if false
```

### &#9989;&nbsp; Question 3.2 (7 Points)  

Define a function which inputs a `radius` (default value `1.0`) and a `color` (default value `"green"`) and plots a circle (centered at the origin) of the given `radius` with the given `color` using Matplotlib. Don't forget to label your axis!

Reminder: The equation of a circle centered at the origin with radius $r$ is $x^2 + y^2 = r^2$. When you solve for $y$ as a function of $x$, there is both the positive and negative square root.

Use this function to plot a green circle of radius 1, an orange circle of radius 2, and a blue circle of radius 3 on the same plot.

*Hint*: It's not necessary, but you can use the line `plt.axis("equal")` to make the horizontal and vertical axes have the same scale. This will make the circles look more... circular.

```python
def circle_plot(radius = 1.0, color = "green"): # defining function that accepts radius and color as parameters, with default arguments
    x_values = np.linspace(-radius, radius, 100) # creating an array of 100 linearly spaced values between negative and positive values of the radius
    y_values = ((radius ** 2) - (x_values ** 2)) ** (1/2) # applying the x values created above to the function for a circle
    y_neg_values = - y_values # accommodating for the negative y values that weren't created above
    plt.plot(x_values, y_values, color, label = "radius " + str(float(radius))) # plot 1: plotting x and positive y values, labeled with the radius length
    plt.plot(x_values, y_neg_values, color) # plot 2: plotting x and negative y values
    plt.axis("equal") # making the scales of the axes equal to avoid distorting the circle
    plt.xlabel("x") # labeling the x axis
    plt.ylabel("y") # labeling the y axis
    plt.title("Plotting Circles") # creating the title
    plt.legend() # invoking a legend using labels specified above
    return

circle_plot() # radius = 1.0, color = green [default]
circle_plot(2, "orange") # radius = 2.0, color = orange
circle_plot(3, "blue") # radius = 3.0, color = blue
```
