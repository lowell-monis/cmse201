# 1. Writing a function (6 points)

In statistics, **arithmatic mean** or just **mean** or **average** is an important factor in calculating the Pearson correlation coefficient (PCC) which is a correlation coefficient that measures linear correlation between two sets of data. The PCC has a value between −1 and 1.

Mathematically,the equation for mean is
### $$\overline{x} =\frac{\sum_{i=1}^N x_i}{N} = \frac{x_1 + x_2 +\cdots + x_N}{N} $$


✅ **Write a function (6 points)**

- Write a function called `average` (**2 points**).
- Your function should take one input called: `x_values` (**2 point**).
- The function should return the mean value of x_values (**2 points**). 

```python
def average(x_values): # defining a function that takes one input. must be sequential data.
    total = 0 # declaring empty variable to store the sum of all x values
    # i will not be using NumPy here, since it is not given whether the input will always be a NumPy array
    for x in x_values:
        total += x # adding all values in x_values to the total using the unary increment assignment operator
    average_x = total / len(x_values) # calculates average outside the loop, dividing total by the number of values
    return average_x # returning the value stored in the average local variable
```

## Part 1.2  [4 points]
    
Using the test data below and the defined function to calculate the average or mean **x_values, y_vaules**. **Demonstrate that your function works by using `np.mean()`**.


```python
import numpy as np
x_values = np.random.normal(1, 2, 200) # 200 sample points from a normal distribution with mean 2 and standard deviation 2.
y_values = np.linspace(1,5,200) # 200 evenly distributed points between 1 and 5.

# the code should run without an error
# i have used np.isclose and the assert statement to verify the two values
assert np.isclose(average(x_values), np.mean(x_values))
assert np.isclose(average(y_values), np.mean(y_values))
```

# 2. Visualizing a model using matplotlib (6 points)

Compound interest is interest accumulated from a principal sum and previously accumulated interest. It is the result of reinvesting or retaining interest that would otherwise be paid out, or of the accumulation of debts from a borrower. The total accumulated value, including the principal sum **P** plus compounded interest **I**, is given by the formula:

$$A= P(1+\frac{r}{n})^{nt}$$
where:

$A$ is the final amount

$P$ is the original principal sum

$r$ is the nominal annual interest rate

$n$ is the compounding frequency

$t$ is the overall length of time the interest is applied (expressed using the same time units as r, usually years).

Suppose an amount of $3,500.00 is deposited in a bank paying an annual interest rate of 4.8%, compounded monthly.

**Use the provided function, model parameters, and array `years`, to calculate a new array of the balance of each year**.

Then, **use matplotlib to make a plot of those balance values as a function of the `years` array.**

To check if the balance values seem plausible, **use the `plt.axvline()` function to make a line at year 16**, which should represent the balance after 16 years. 

**You should also make sure to label your x and y axes.**


```python
# Will need to use NumPy for the function
import numpy as np
# Will need to import matplotlib for the plotting
import matplotlib.pyplot as plt
%matplotlib inline

# This function calculates the total accumulated value over a range of years
def total_accumulated_value(t, P, r, n):
     return P*(1 + r/n)**(n*t)

# Create year array
years = np.arange(0,35,1) + 1
```

```python
balance = total_accumulated_value(years, 3500, 0.048, 12) # creating array for account balances

plt.plot(years, balance, "green")
plt.axvline(16, color = "red")
plt.xlabel("Year")
plt.ylabel("Account Balance")
plt.title("Yearly Account Balance when Interest is Compounded Annually; Principal of $3,500,00 @ 4.8% monthly")
```

# 3. Loading a dataset with NumPy and visualizing it (10 points)

The file, `Boston.csv` consists of housing values (in thousands of dollars) and other information about Boston census tracts. Suppose that you are hired as a data scientist in an investing company, you will use this data to visualize relations between medv (median house value), rm (average number of rooms per house), and age (proportion of owner-occupied units built before 1940).

First, **Using numpy (not pandas!), load the `Boston.csv` file into your notebook.** Inspect the data file and ensure that you only read in the columns you are interested in for full credit. (If you can't figure out how to use NumPy, you can use Pandas for partial credit). Then, **draw simple _scatter plots_ between medv, rm, and age** (one plot for medv and rm, and another one for medv and age).


**Make sure you label your x and y axes.** Also, if you can't get the data loaded, you can still put your plot commands as you would have done if you had been successful (for partial credit).

```python
# Put your code here
boston = np.loadtxt("Boston.csv", delimiter = ",", skiprows = 1) # loading data, skipping first row
medv = np.array([]) # empty arrays to load data into
rm = np.array([])
age = np.array([])

# using a loop and np.append to add individual values to the arrays
for data in boston:
    medv = np.append(medv, data[-1])
    rm = np.append(rm, data[5])
    age = np.append(age, data[6])

# plot
plt.figure(figsize=(10,5))
plt.suptitle("Comparisons concerning Median House Value") # main title

plt.subplot(1,2,1)
plt.scatter(rm, medv, color = "red", s = 2)
plt.title("vs. Rooms per House")
plt.xlabel("Average number of rooms in the house")
plt.ylabel("Median House Value")

plt.subplot(1,2,2)
plt.scatter(age, medv, color = "green", s = 2)
plt.title("vs. Proportion of Pre-1940 Units")
plt.xlabel("Proportion of Owner-Occupied units built before 1940")
plt.ylabel("Median House Value")

plt.tight_layout()
```
